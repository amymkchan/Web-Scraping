# ministore_WebScrap & data clean vodka

```python
from selenium import webdriver  
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
```

```python
#Define Variable
extract_product="alcohol"
category="vodka"

OutProductsFile = "Input/ministore_ProdInfo_"+extract_product+"_"+category+"_"+time.strftime("%Y%m%d")+".csv"
OutProductsFileClean = "Output/ministore_ProdInfo_"+extract_product+"_"+category+"_"+time.strftime("%Y%m%d")+"_clean.csv"
#print (OutFile)

# set the URL for web scraping
url="https://ministore.com.hk/collections/grains-vodka" 

# The chrome browser is saved in /usr/local/bin
#driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
# datasets
descriptions=[]
prices = []
originals=[]
stock_statuses=[]
URLs=[]
```

```python
# for incognito/private mode
#chrome_options = webdriver.ChromeOptions()
#chrome_options.add_argument("--incognito")

# The chrome browser is saved in /usr/local/bin
#driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)

# Not in incognito/private mode 
driver = webdriver.Chrome(ChromeDriverManager().install())

driver.get(url) # go to the URL
time.sleep(5) # time.sleep(T) means wait for T seconds for the chrome browser to load the URL page
```

```python
for i in range(1,1000):
    # description
    # if description = null, then stop 
    web_description = "//div[@class='grid-uniform'][1]/div["+str(i)+"]/a[1]/p"
    elem = driver.find_elements_by_xpath(web_description)
    #print (elem)
    if len(elem) > 0:
        description = driver.find_element_by_xpath(web_description) 
        #print(description.text)
        descriptions.append(description.text)
        
        # URLs
        web_URL ="//div[@class='grid-uniform'][1]/div["+str(i)+"]/a[1]"
        ids=driver.find_elements_by_xpath(web_URL)
        for ii in ids:
            #print(ii.get_attribute("href"))
            URLs.append(ii.get_attribute("href"))
            
        #stock_statuses
        web_status = "//div[@class='grid-uniform'][1]/div["+str(i)+"]/a[1]/div[1]/div[1]/div[1]"
        elem = driver.find_elements_by_xpath(web_status)
        if len(elem) > 0:  #found the element
            # handle sold out
            status = driver.find_element_by_xpath(web_status) 
            #print(status.text)
            stock_statuses.append(status.text)
        else:
            stock_statuses.append("")
            
        #price
        web_price = "//div[@class='grid-uniform'][1]/div["+str(i)+"]/a[1]//span[2]"
        elem = driver.find_elements_by_xpath(web_price)
        if len(elem) > 0:  #found the element
            # handle the price
            price = driver.find_element_by_xpath(web_price) 
            #print(price.text)
            prices.append(price.text)
        else:
            prices.append("")
        
        #original
        web_original = "//div[@class='grid-uniform'][1]/div["+str(i)+"]/a[1]//small[2]//span[2]"
        elem = driver.find_elements_by_xpath(web_original)
        if len(elem) > 0:  #found the element
            # handle the price
            original = driver.find_element_by_xpath(web_original) 
            #print(origial.text)
            originals.append(original.text)
        else:
            originals.append("")
        
    else:
         break    # break here
    
```

```python
# Group the lists to dataframe
ministore_dataset = pd.DataFrame({
     'co': 'ministore',
     'category':category,
     'extract_date': time.strftime("%Y%m%d"),
     'description': descriptions,
     'price': prices,
     'original': originals,
     'stock_status': stock_statuses,
     'URL': URLs
})
```

# Data clean
- price: remove ['$', 'HKD', ',']

- original: remove ['$', 'HKD', ',']

- replace price with original if original is not null

```python
# copy to clean dataset
df_clean=ministore_dataset.copy()

# price : remove{'$',',','HKD')
df_clean['price'] = df_clean.price.str.replace('HKD' , '')
df_clean['price'] = df_clean.price.str.replace('$' , '')
df_clean['price'] = df_clean.price.str.replace(',' , '')

# original : remove{'$',',','HKD'))
df_clean['original'] = df_clean.original.str.replace('HKD' , '')
df_clean['original'] = df_clean.original.str.replace('$' , '')
df_clean['original'] = df_clean.original.str.replace(',' , '')

# - Price with original if original is not null
for i in range(0,len(df_clean)):
    #print(i,df_clean.original[i])
    if df_clean.original[i]=='':
    #    if pd.isna(df_clean.original[i]):
        #print (df_clean.price[i])
        df_clean.original[i]=df_clean.price[i]
        df_clean.price[i]=0

```

```python
# set the "price" and "promotion" column to float
df_clean = df_clean.astype({'price': float})
df_clean = df_clean.astype({'original': float})

# take a look on the dataset's info and attributes after changes
df_clean.info() 
```
