# Web Scraping - HKTVMall  伏特加

```python
from selenium import webdriver  
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
```

```python
#Define Variable
extract_product="alcohol"
category="vodka"

# outfile
OutProductsFile = "Input/HKTVMall_ProdInfo_"+extract_product+"_"+category+"_"+time.strftime("%Y%m%d")+".csv"
#print (OutProductsFile)

# set the URL for web scraping
url="https://www.hktvmall.com/hktv/zh/main/search?page=0&q=%3Arelevance%3Acategory%3AAA11306030001%3Azone%3Asupermarket%3Astreet%3Amain%3A"
#url(Eng)="https://www.hktvmall.com/hktv/en/main/search?page=0&q=%3Arelevance%3Acategory%3AAA11306030001%3Azone%3Asupermarket%3Astreet%3Amain%3A"
# The chrome browser is saved in /usr/local/bin
#driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
# datasets
band_names=[]
weights=[]
prices = []
promotions=[]
stock_statuses=[]
URLs=[]
```

```python
# for incognito/private mode
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")

# The chrome browser is saved in /usr/local/bin
driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)

# Not in incognito/private mode 
#driver = webdriver.Chrome(ChromeDriverManager().install())

driver.get(url) # go to the URL
time.sleep(5) # time.sleep(T) means wait for T seconds for the chrome browser to load the URL page
```

### start to web scraping

```python
# get the last record no of option list
last_option = driver.find_element_by_xpath("//*[@class='page-nav']/select[1]/option[last()]")
#print(last_option.text)
last_no=int(last_option.text)+1
#print(last_no)

# start extracting
for i in range(1,last_no): # range(X,Y) means loop the page from X to Y-1 
#for i in range(3,last_no): # range(X,Y) means loop the page from X to Y-1 
    #print(i)
    link = driver.find_element_by_xpath("//*[@class='page-nav']/select[1]/option["+str(i)+"]")
    link.click() # click on "Next Page" link
    time.sleep(5) # time.sleep(T) means wait for T seconds for the chrome browser to load the URL page
    
    for j in range(1,70):
        #print (i,j)    
        web_band_name = "//*[contains(@class,'product-brief-wrapper')]["+str(j)+"]/div[1]/div[2]/div[1]/div[1]"
        elem = driver.find_elements_by_xpath(web_band_name)
        if len(elem) > 0:  #found the element
            # advertising
            web_advFlag = "//*[contains(@class,'product-brief-wrapper')]["+str(j)+"]/div[1]/div[1]/p[1]"
            elem = driver.find_elements_by_xpath(web_advFlag)
            if len(elem) == 0:  # not advertising
                # band_name
                band_name = driver.find_element_by_xpath(web_band_name) 
                #print(band_name.text)
                band_names.append(band_name.text)
            
                # weight
                web_weight = "//*[contains(@class,'product-brief-wrapper')]["+str(j)+"]/div[1]/div[2]/div[1]/div[2]"
                elem = driver.find_elements_by_xpath(web_weight)
                if len(elem) > 0:  #found the element
                    weight = driver.find_element_by_xpath(web_weight) 
                    #print(weight.text)
                    weights.append(weight.text)
                else:
                    weights.append("")
                
                # price 
                web_price = "//*[contains(@class,'product-brief-wrapper')]["+str(j)+"]/div[1]/div[2]/div[2]/div[1]/div[1]"
                elem = driver.find_elements_by_xpath(web_price)
                if len(elem) > 0:  #found the element
                    price = driver.find_element_by_xpath(web_price) 
                    #print(j, price.text)
                    if len(price.text) ==0:
                        prices.append("$0")
                    else:
                        prices.append(price.text)
                else:
                    prices.append("$0")
                    
                # promotion 
                web_promotion = "//*[contains(@class,'product-brief-wrapper')]["+str(j)+"]/div[1]/div[2]/div[2]/div[1]/div[2]"
                elem = driver.find_elements_by_xpath(web_promotion)
                if len(elem) > 0:  #found the element
                    promotion = driver.find_element_by_xpath(web_promotion) 
                    #print(promotion.text)
                    promotions.append(promotion.text)
                else:
                    promotions.append("")

                # stock_status 
                web_status = "//*[contains(@class,'product-brief-wrapper')]["+str(j)+"]/div[1]/div[2]/div[3]"
                elem = driver.find_elements_by_xpath( web_status)
                if len(elem) > 0:  #found the element
                    status = driver.find_element_by_xpath(web_status)
                    txt_status="stock on hand"
                    if  status.text=="有貨通知我":
                        #print(status.text)
                        txt_status="out of stock"
                    stock_statuses.append(txt_status)
                else:
                    stock_statuses.append("Not found")

                # URL 
                web_URL = "//*[contains(@class,'product-brief-wrapper')]["+str(j)+"]/div[1]/a[1]"
                ids=driver.find_elements_by_xpath(web_URL)
                for ii in ids:
                    URLs.append(ii.get_attribute("href"))  
        else:
            #skip the loop j 
            break    # break here
    
    # exit loop j 
```

```python
# Group the lists to dataframe
HKTVMall_dataset  = pd.DataFrame({
     'co': 'HKTVMall',
     'category':category,
     'extract_date': time.strftime("%Y%m%d"),
     'band': band_names,
     'weight': weights,
     'price': prices,
     'promotion': promotions,
     'stock_status': stock_statuses,
     'URL': URLs
})
```
