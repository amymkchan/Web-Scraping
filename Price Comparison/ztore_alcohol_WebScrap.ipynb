# Web Scraping - ztore 伏特加

```python
from selenium import webdriver  
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
```

```python
#Define Variable
extract_product="alcohol"
category="vodka"

#OutFile = "output/zstoreHref_"+time.strftime("%Y%m%d-%H%M%S")+".csv"
OutFile = "Input/zstore_Href_"+extract_product+"_"+category+"_"+time.strftime("%Y%m%d")+".csv"
OutProductsFile = "Input/zstore_ProdInfo_"+extract_product+"_"+category+"_"+time.strftime("%Y%m%d")+".csv"
#print (OutFile)

# set the URL for web scraping
url="https://www.ztore.com/tc/category/all/alcohol/liquor-spirits/vodka" 

# The chrome browser is saved in /usr/local/bin
#driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
# initialize the lists for storing data
Hreflinks=[]
```

```python
# datasets
band_names=[]
descriptions=[]
volumes=[]
weights=[]
prices = []
originals=[]
promotions=[]
stock_statuses=[]
countries=[] 
URLs=[]
```

```python
# for incognito/private mode
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")

# The chrome browser is saved in /usr/local/bin
driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)

driver.get(url) # go to the URL
time.sleep(5) # time.sleep(T) means wait for T seconds for the chrome browser to load the URL page
```

### start to web scraping

```python
# click "view All" Button
web_link="//*[@class='viewAllButton']"
elem = driver.find_elements_by_xpath(web_link)
if len(elem) > 0:  #found the element
    link = driver.find_element_by_xpath(web_link)
    link.click() # click on "view All" link
#time.sleep(1) # time.sleep(T) means wait for T seconds for the chrome browser to load the URL page
```

```python
#Scroll to Top
driver.execute_script("scroll(0, 0)")
```

```python
if len(elem) > 0:  #found the element
    for i in range(150): # adjust integer value for need
        # change right side number for scroll convenience or destination
        driver.execute_script("window.scrollBy(0, 250)")
        # you can change time integer to float or remove
        time.sleep(1)
```

```python
#select ALL products URL
xpath="//*[contains(@class,'jsx-638078379 jsx-1176293161 label-ctnr')]/a[1]"

ids=driver.find_elements_by_xpath(xpath)
for ii in ids:
    print(ii.get_attribute("href"))
    Hreflinks.append(ii.get_attribute("href"))
```

```python
# Link info to dataframe
zstore_Href_df = pd.DataFrame({
     'HRef': Hreflinks 
})
```

```python
zstore_Href_df_Sort=zstore_Href_df[['HRef']].drop_duplicates()
zstore_Href_df_Sort
```

```python
# export the dataframe to csv file (1st time)
zstore_Href_df_Sort.to_csv(OutFile, index=False) 
```

```python
#Start extract all data
for i in range(0,len(zstore_Href_df_Sort)):
#for i in range(544,len(zstore_Href_df_Sort)):
    #print (zstore_Href_df_Sort.HRef[i])
    driver.get(zstore_Href_df_Sort.HRef[i]) # go to the URL
    URLs.append(zstore_Href_df_Sort.HRef[i])
    
    #band
    web_band_name = "//*[contains(@class,'brand')]"
    elem = driver.find_elements_by_xpath(web_band_name)
    if len(elem) > 0:  #found the element
        band_name = driver.find_element_by_xpath(web_band_name) 
        #print(band_name.text)
        band_names.append(band_name.text)
    else:
        band_names.append("")
    
    #description
    web_description = "//*[@class='jsx-2308771206']"
    elem = driver.find_elements_by_xpath(web_description)
    if len(elem) > 0:  #found the element
        description = driver.find_element_by_xpath(web_description) 
        #print(description.text)
        descriptions.append(description.text)
    else:
        descriptions.append("")

    #volumes
    web_volume = "//*[@class='jsx-2308771206 flex-center']/div[1]/span[1]"
    elem = driver.find_elements_by_xpath(web_volume)
    if len(elem) > 0:  #found the element
        volume = driver.find_element_by_xpath(web_volume) 
        #print(volume.text)
        volumes.append(volume.text)
    else:
        volumes.append("")
        
    #weights
    web_weight = "//*[@class='jsx-2308771206 flex-center']/div[1]/span[2]"
    elem = driver.find_elements_by_xpath(web_weight)
    if len(elem) > 0:  #found the element
        weight = driver.find_element_by_xpath(web_weight) 
        #print(weight.text)
        weights.append(weight.text)
    else:
        weights.append("")

    #prices
    web_price = "//*[@class='jsx-2308771206 price']"
    elem = driver.find_elements_by_xpath(web_price)
    if len(elem) > 0:  #found the element
        price = driver.find_element_by_xpath(web_price) 
        #print(price.text)
        prices.append(price.text)
    else:
        prices.append("")

    #originals 
    web_original = "//*[@class='jsx-2308771206 price']/div[1]/span[2]"
    elem = driver.find_elements_by_xpath(web_original)
    if len(elem) > 0:  #found the element
        original = driver.find_element_by_xpath(web_original) 
        #print(original.text)
        originals.append(original.text)
    else:
        originals.append("")

    #promotions
    web_promotion = "//*[@class='jsx-2308771206 price']/div[1]/span[1]"
    elem = driver.find_elements_by_xpath(web_promotion)
    if len(elem) > 0:  #found the element
        promotion = driver.find_element_by_xpath(web_promotion) 
        #print(promotion.text)
        promotions.append(promotion.text)
    else:
        promotions.append("")

    #stock_statuses
    web_status = "//*[@class='jsx-2308771206 add-to-cart-fav-wrapper']/div[1]"
    status = driver.find_element_by_xpath(web_status)
    elem = driver.find_elements_by_xpath(web_status)
    if len(elem) > 0:  #found the element
        txt_status="stock on hand"
        if  status.text=="返貨通知":
            #print(status.text)
            txt_status="out of stock"
        stock_statuses.append(txt_status)
    else:
        stock_statuses.append("Not found")        

    #country=[] 
    #web_country = "//*[@class='jsx-2308771206 info-row']/div[2]"
    web_country = "//*[@class='jsx-2308771206 info-row-country']"
    elem = driver.find_elements_by_xpath(web_country)
    if len(elem) > 0:  #found the element
        country = driver.find_element_by_xpath(web_country) 
        #print(country.text)
        countries.append(country.text)
    else:
        countries.append("")
   
    #autorefresh chrome
    driver.refresh()
```

```python
# Group the lists to dataframe
zstore_dataset = pd.DataFrame({
     'co': 'zstore',
     'category':category,
     'extract_date': time.strftime("%Y%m%d"),
     'band': band_names,
     'description': descriptions,
     'volume': volumes,
     'weight': weights,
     'price': prices,
     'original': originals,
     'promotion': promotions,
     'stock_status': stock_statuses,
     'country': countries,  
     'URL': URLs
})
```
